### OLS (Ordinary least squares)

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,  r2_score, mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression
```

###### only continous features

```python
import statsmodels.api as sm

X = df_amount.drop([ "WrittenSales"],axis=1)
y = df_amount["WrittenSales"]

# Note the difference in argument order
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model

print('R2:', r2_score(y, predictions))
print('MAE:', mean_absolute_error(y, predictions))
print('MSE:', mean_squared_error(y, predictions))
print('RMSE:',np.sqrt(mean_squared_error(y, predictions)))

# Print out the statistics
model.summary()
```

###### only continous features  - train/test

```python
X = df_amount.drop([ "WrittenSales"],axis=1)
y = df_amount["WrittenSales"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)

# Note the difference in argument order
model = sm.OLS(y_train, X_train).fit()
y_pred = model.predict(X_test) # make the predictions by the model

print('R2:', r2_score(y_test,y_pred))
print('MAE:', mean_absolute_error(y_test,y_pred))
print('MSE:', mean_squared_error(y_test,y_pred))
print('RMSE:',np.sqrt(mean_squared_error(y_test,y_pred)))

# Print out the statistics
model.summary()
```

### all features - 10-fold cross validation 

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
X = data.iloc[:, 1:-4].values
y = data.Traffic.values.reshape(-1,1)

lm = LinearRegression()
model = lm.fit(X, y)
crossvalidation = KFold(n_splits=5, random_state=None, shuffle=False)

mse_scores = cross_val_score(model, X, y, scoring="neg_mean_squared_error", cv=crossvalidation, n_jobs=1)
mae_scores = cross_val_score(model, X, y, scoring="neg_mean_absolute_error", cv=crossvalidation, n_jobs=1)
R2_scores = cross_val_score(model, X, y, scoring='r2', cv=crossvalidation)

print("R2:", str(np.mean(np.abs(R2_scores))))
print("MAE:", str(np.mean(np.abs(mae_scores))))
print("MSE:", str(np.mean(np.abs(mse_scores))))
print("RMSE:",str(np.sqrt(np.mean(np.abs(mse_scores)))))
```
